```python
results = []
fpr = []
overlapping = []
alpha=0.95
namedict = {"calPredCorrect":"CalPred","PredInterval":"PredInterval"}
for h2 in [0.2,0.5,0.8]:
    for causal_percent in [0.001,0.01,0.1,1]:
        print(h2,causal_percent)
        for i in range(30):

            
            df = pd.read_csv(f"sim_df_PGS.h2{h2}.causal{causal_percent}.seed{i}.tsv",sep="\t")
            t = np.quantile(df.iloc[:50000].y,0.95)
            for scheme in ["calPredCorrect","PredInterval"]:
                threshold = t
                pred = pd.read_csv(f"{scheme}.alpha{alpha}.h2{h2}.causal{causal_percent}.seed{i}.pred.tsv",sep="\t")
                results.append([h2,causal_percent,namedict[scheme],i,((pred.y>=threshold) & (pred.upper>=threshold)).sum()/(pred.y>=threshold).sum()*100,50000])

                values = []
                for c in [0.05,0.1,0.25,0.5,0.95]:
                    lt = np.quantile(df.iloc[:50000].y,c)
                    values.append(((pred.y<=lt) & (pred.upper>=threshold)).sum()/(pred.y<=lt).sum()*100)
                fpr.append([h2,causal_percent,namedict[scheme],i]+values+[50000])

            df = pd.read_csv(f"sim_5000/sim_df_PGS.nc.h2{h2}.causal{causal_percent}.seed{i}.tsv",sep="\t")
            t = np.quantile(df.iloc[:5000].y,0.95)
            for scheme in ["calPredCorrect","PredInterval"]:
                threshold = t
                pred = pd.read_csv(f"sim_5000/{scheme}.alpha{alpha}.h2{h2}.causal{causal_percent}.seed{i}.pred.tsv",sep="\t")
                results.append([h2,causal_percent,namedict[scheme],i,((pred.y>=threshold) & (pred.upper>=threshold)).sum()/(pred.y>=threshold).sum()*100,5000])
                
                values = []
                for c in [0.05,0.1,0.25,0.5,0.95]:
                    lt = np.quantile(df.iloc[:5000].y,c)
                    values.append(((pred.y<=lt) & (pred.upper>=threshold)).sum()/(pred.y<=lt).sum()*100)
                fpr.append([h2,causal_percent,namedict[scheme],i]+values+[5000])

                


results = pd.DataFrame(results)
fpr = pd.DataFrame(fpr)
```
